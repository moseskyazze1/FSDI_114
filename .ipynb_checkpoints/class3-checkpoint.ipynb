```python
#from scratch implementation

class SinglyLinkedList:
    class __Node:
        def __init__(self, data):
            self.data = data
            self.next = None

    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, value):
        new_node = self.__Node(value)
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
            
    #INSERT IMPLEMENTATION: 
    #create new node
    #insert new node at the head
    #let our index be 0 (first position)
    #insert new node at the point of the current node in head
    def insert(self, index, value):
        new_node = self.__Node(value)
        if index == 0:
            new_node.next = self.head
            self.head = new_node
            if self.tail is None:  #checking whether the list is empty
                self.tail = new_node  #self.tail is now the new node since its both head and tail
            return

        current = self.head  #starts at the head of the list.
        previous = None    #points to the node before we insert new node
        count = 0    #this is tracking the current index during the whole operation

        while count < index and current:  #the loop moves through list until current becomes none
            previous = current  #updating previous node with current node
            current = current.next  #updated to point to the node in the list
            count += 1  #increasing the count

    #REMOVE IMPLEMENTATION
    #replace the head with the current node
    #check and remove the head node 
    #update the head node to be the node following the cureent head that is introduced
    #make sure thta the list is empty 
    #travel across the list to find the target node and set previous node to current node
    #replace the current node with the next node since its replaced with previous node
    #check whether the value is present in the list
    #remove the node that has no next node ie if tail has no next node then remove it and set tail to previous

     def remove(self, value):
        current = self.head
        previous = None

        if current and current.data == value:  # Check if the head node needs to be removed
            self.head = current.next
            if self.head is None:  # If list becomes empty
                self.tail = None
            return

        while current and current.data != value:  # Traverse the list to find the node to remove
            previous = current
            current = current.next

        if current is None:  # Value not found
            raise ValueError("Value not found in the list")

        previous.next = current.next  # Remove the node
        if current.next is None:  # If it was the tail
            self.tail = previous


    def __str__(self):
        out = "["
        current = self.head
        if current:
            out += "%s" % repr(current.data)
            while current.next:
                current = current.next
                out += ", %s" % repr(current.data)
        out += "]"
        return out

        
        

```

```python
sll = SinglyLinkedList()
print(sll)

sll.append(0)
print(sll)

sll.append(1)
print(sll)
```
